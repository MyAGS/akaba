```python
#Melakukan Cleaning
meanTotal = df['Total'].mean()
stdHP = df['HP'].std()
meanTotal = df['Total'].mean()
stdHP = df['HP'].std()

dataClean = df[(np.abs((df['Total']-meanTotal)/stdTotal) <= 3) 
                 & (np.abs((df['HP']-meanHP)/stdHP) <= 3)]
```

```python
# scatter plot setelah menghilangkan outliers

import matplotlib.pyplot as plt
plt.scatter(dataClean['Total'], dataClean['HP'])
plt.xlabel('Total')
plt.ylabel('HP')
plt.show()
```

```python
cdf = dataClean[['Total','HP']]
cdf.head(9)
```

```python
msk = np.random.rand(len(dataClean)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
#mencari nilai Coefficients dan intercept
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['Total']])
train_y = np.asanyarray(train[['HP']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
#bentuk hasil 
plt.scatter(train.Total, train.HP,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Total")
plt.ylabel("HP")
```

```python
#mencari mae, MSE, R2
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['Total']])
test_y = np.asanyarray(test[['HP']])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```

```python
#Menggunakan scatter mattrik untuk langkas klasifikasinya
from pandas.plotting import scatter_matrix #aslinya: pandas.tools.plotting
from matplotlib import cm
feature_names = ['Total','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']
X = dataClean[feature_names]
y = dataClean['Nomer']
cmap = cm.get_cmap('gnuplot')
scatter = scatter_matrix(X, c = y, marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(9,9), cmap = cmap)
plt.suptitle('Scatter-matrix for each input variable')
plt.savefig('pokemon_scatter_matrix')
```

```python
#maaf buat apa saya tidak tahu :V
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#model algoritma logistic regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(logreg.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(logreg.score(X_test, y_test)))
```

```python
#model algoritma decision tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_train, y_train)
print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf.score(X_train, y_train)))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf.score(X_test, y_test)))
```

```python
#memprediksi jenis pokemon menggunakan data Total	HP	Attack	Defense	Sp. Atk	Sp. Def	Speed
pokemon_predictionlog = clf.predict(scaler.transform([[300,60,62,63,80,80,60]]))
look_up_Name = dict(zip(dataClean.Nomer.unique(),dataClean.Name.unique()))
print('Pokemon diprediksi bernama :',look_up_Name[pokemon_predictionlog[0]])
```
